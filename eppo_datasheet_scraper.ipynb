```python
"""
scrape urls in dfs.Datasheet and parse 'datasheet' div class, put the content in a nested dict of this type {<h3>:[<h4>:<p>]}. save the dict in a new dict with the second last word of the url as key and the dict as value 
Example:
"""
a = """
<h1 class="title">EPPO Datasheet: <i>Palm lethal yellowing type syndromes</i></h1>
<!-- -->
<h3>IDENTITY</h3>
<div class="para0">
    <strong>Preferred name:</strong> <i>Palm lethal yellowing type syndromes</i><br>
        <strong>Taxonomic position: </strong>Bacteria: Tenericutes: Mollicutes: Acholeplasmatales: Acholeplasmataceae<br>
	<strong>Other scientific names: </strong> <i>Coconut lethal yellowing phytoplasma</i>, <i>Palm lethal yellowing disease</i><br>
	<strong>Common names in English: </strong> Awka wilt disease, Bogia coconut syndrome, Cape St Paul wilt of coconut, Palm lethal yellowing disease, bronze leaf wilt of coconut (JM), kaincope disease of coconut, kribi disease of coconut, lethal yellowing of coconut, lethal yellowing syndromes of coconut, lethal yellowing type syndromes<br>
<a href="https://gd.eppo.int/taxon/PHYP56/">view more common names online...</a><br>
<!-- -->
<strong>Notes on taxonomy and nomenclature</strong>
<div class="para0">
<p style="text-align: justify;">Lethal yellowing disease of coconut was first observed in the Caribbean in the 19<sup>th</sup> century. However, it was only in 1972 that it was confirmed to be caused by phytoplasmas (Beakbane <em>et al.,</em> 1972; Plavsic-Banjac <em>et al.,</em> 1972). The many common names cited above, often from the affected localities, have been used for diseases of coconuts which are all now considered to be caused by different phytoplasmas associated with common syndromes: Lethal Yellowing Type Syndromes. These diseases differ mainly in their geographic distributions. Their host ranges all contain plants from the Arecacae family, including the coconut tree (<em>Cocos nucifera</em>). Many coconut diseases of unknown aetiology are described in the older literature and records of these have been appraised by Howard (1983). The name ‘bronze leaf wilt’ has been used for lethal yellowing in Jamaica but corresponds to a different disease in South America.</p><p style="text-align: justify;">Because no complete genome sequences are currently available for any of these phytoplasmas, their taxonomy is based on 16S rRNA sequences, and two parallel classification systems have been developed, the 16Sr group system, based on restriction enzyme digest profiles of the 16S rDNA, and the <em>‘Candidatus</em> Phytoplasma’ species system, in which phytoplasmas sharing less than 97.5% similarity of their 16S rRNA gene sequence may be ascribed to different ‘<em>Ca</em>. Phytoplasma’ species when they are characterised by distinctive biological, phytopathological and genetic properties (Bertaccini <em>et al</em>., 2022, EFSA, 2017).</p><p style="text-align: justify;">Phytoplasmas associated with palm lethal yellowing type syndromes are placed in the 16SrIV and 16SrXXII groups, both of which are divided in different subgroups reflecting the restriction profile of the 16SrRNA gene (Gurr <em>et al</em>., 2016; Wei <em>et al</em>., 2007).</p><p style="text-align: justify;">In the framework of this datasheet, the following ‘<em>Candidatus</em> Phytoplasma’ species are considered to be associated with palm lethal yellowing type syndromes:</p><ul><li>‘<em>Candidatus</em> Phytoplasma aculeata’</li><li>‘<em>Candidatus</em> Phytoplasma cocostanzaniae’</li><li>‘<em>Candidatus</em> Phytoplasma dypsidis'</li><li><em>'Candidatus</em> Phytoplasma hispanola'</li><li><em>'Candidatus</em> Phytoplasma noviguineense'</li><li>‘<em>Candidatus</em> Phytoplasma palmae’</li><li>‘<em>Candidatus</em> Phytoplasma palmicola’</li><li><br></li></ul></div>
	<strong>EPPO Categorization:</strong> A1 list<br>
	<strong>EU Categorization:</strong> A1 Quarantine pest (Annex II A)<br>
	  <a href="https://gd.eppo.int/taxon/PHYP56/categorization">view more categorizations online...</a><br>

  <strong>EPPO Code:</strong> PHYP56</div>

    <!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> HOSTS<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1" style="display: none;">
<p style="text-align: justify;">The main host is coconuts (<em>Cocos nucifera</em>) but the lethal yellowing type syndromes have also been found on many different palms (Arecaceae)<em>.</em> The occurrence of symptoms similar to lethal yellowing together with the presence of phytoplasmas among collections of palms in affected areas in Florida (USA) indicates that more than 30 species of palms are susceptible to infection. Palm lethal yellowing type syndrome phytoplasmas have been detected in non Arecaceae plants, including different grasses in plots where palms were infested (Arocha-Rosete <em>et al.</em>, 2016; Gurr <em>et al</em>., 2016). The relevance of these plants as potential reservoirs of the phytoplasmas is not clear (EFSA, 2017). <em>Ca</em> ‘Phytoplasma noviguineense’ also includes banana (<em>Musa</em> sp. ) as another economically important host crop.</p><p>A detailed list of host plants for each phytoplasma species is available in EPPO Global Database. A global host list for all host plants of palm lethal yellowing type syndromes phytoplasmas is as follows:</p><strong>Host list:</strong> <i>Acrocomia aculeata</i>, <i>Adonidia merrillii</i>, <i>Aiphanes horrida</i>, <i>Aiphanes minima</i>, <i>Allagoptera arenaria</i>, <i>Areca catechu</i>, <i>Arecaceae</i>, <i>Arenga engleri</i>, <i>Arenga pinnata</i>, <i>Arenga</i>, <i>Attalea butyracea</i>, <i>Bismarckia nobilis</i>, <i>Borassus flabellifer</i>, <i>Brahea brandegeei</i>, <i>Butia odorata</i>, <i>Carludovica palmata</i>, <i>Carpentaria acuminata</i>, <i>Caryota mitis</i>, <i>Caryota rumphiana</i>, <i>Caryota urens</i>, <i>Chelyocarpus chuco</i>, <i>Chrysalidocarpus cabadae</i>, <i>Chrysalidocarpus decaryi</i>, <i>Chrysalidocarpus leptocheilos</i>, <i>Chrysalidocarpus lutescens</i>, <i>Cleome rutidosperma</i>, <i>Cocos nucifera</i>, <i>Cocos</i>, <i>Corypha umbraculifera</i>, <i>Corypha utan</i>, <i>Corypha</i>, <i>Cryosophila warscewiczii</i>, <i>Cyanthillium cinereum</i>, <i>Dictyosperma album</i>, <i>Dictyosperma</i>, <i>Elaeis guineensis</i>, <i>Gaussia attenuata</i>, <i>Gaussia</i>, <i>Howea belmoreana</i>, <i>Howea forsteriana</i>, <i>Hyophorbe verschaffeltii</i>, <i>Hyophorbe</i>, <i>Latania lontaroides</i>, <i>Livistona australis</i>, <i>Livistona chinensis</i>, <i>Macroptilium lathyroides</i>, <i>Nannorrhops ritchieana</i>, <i>Phoenix canariensis</i>, <i>Phoenix dactylifera</i>, <i>Phoenix reclinata</i>, <i>Phoenix roebelenii</i>, <i>Phoenix sylvestris</i>, <i>Phoenix</i>, <i>Pritchardia pacifica</i>, <i>Pritchardia thurstonii</i>, <i>Pseudophoenix sargentii</i>, <i>Ravenea rivularis</i>, <i>Roystonea regia</i>, <i>Sabal mexicana</i>, <i>Sabal palmetto</i>, <i>Saribus rotundifolius</i>, <i>Stachytarpheta jamaicensis</i>, <i>Syagrus romanzoffiana</i>, <i>Syagrus schizophylla</i>, <i>Syagrus sp.</i>, <i>Synedrella nodiflora</i>, <i>Trachycarpus fortunei</i>, <i>Veitchia arecina</i>, <i>Washingtonia robusta</i>
</div>
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> GEOGRAPHICAL DISTRIBUTION<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1" "="">
<p style="text-align: justify;">Lethal yellowing type syndromes are widely distributed in the intertropical zone. They are present in North and Central America and the Caribbean (‘<em>Ca</em>. Phytoplasma palmae’ and ‘<em>Ca</em>. Phytoplasma aculeata’ and '<em>Ca</em>. Phytoplasma hispanola'), in Africa (‘<em>Ca</em>. Phytoplasma palmicola’ and ‘<em>Ca</em>. Phytoplasma cocostanzaniae’ and in Oceania (<em>'Ca</em>. Phytoplasma noviguineense'<u>,</u> and ‘<em>Ca</em>. Phytoplasma dypsidis').</p><p>Distribution maps for each phytoplasma species is available in EPPO Global Database. The global distribution of lethal yellowing type syndromes of palms is as follows:</p><img class="img-responsive" src="/taxon/PHYP56/download/distributionmap_png" style="margin:10px 0px 10px 0px">
<strong>Africa: </strong>Cameroon, Cote d'Ivoire, Equatorial Guinea, Ghana, Kenya, Madagascar, Mozambique, Nigeria, Tanzania, Togo<br>
<strong>North America: </strong>Mexico, United States of America (Florida, Louisiana, Texas)<br>
<strong>Central America and Caribbean: </strong>Antigua and Barbuda, Belize, Cayman Islands, Cuba, Dominican Republic, Guadeloupe, Guatemala, Haiti, Honduras, Jamaica, Netherlands Antilles, St Kitts-Nevis<br>
<strong>Oceania: </strong>Australia (Queensland), Papua New Guinea<br>
</div>
<!-- -->
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> BIOLOGY<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1">
<p style="text-align: justify;">Palm lethal yellowing type syndromes are associated with phytoplasmas. Mycoplasma-like particles have been found in the sieve-tube elements of the phloem of coconut and other palms exhibiting characteristic symptoms (Beakbane <em>et al.,</em> 1972; Plavsic-Banjac <em>et al.,</em> 1972). Oxytetracycline treatment causes remission of disease symptoms (McCoy, 1975). Since then, the techniques of molecular biology and DNA sequencing have made it possible to show the association of phytoplasmas with these syndromes (Gurr <em>et al.</em>, 2016; Jones <em>et al</em>., 2021; Miyazaki <em>et al</em>., 2018; Soto <em>et al</em>., 2021).</p><p style="text-align: justify;">Marked differences in susceptibility to the phytoplasmas are reported in different varieties of coconut palms as well as between other palm species. Susceptibility also varies with the phytoplasma species involved (Howard <em>et al</em>., 1979; Gurr <em>et al</em>., 2016; Palma-Cancino <em>et al.</em>, 2023).</p><p style="text-align: justify;">To date, the only insect identified as a vector of a palm lethal yellowing type syndromes -associated phytoplasma is <em>Haplaxius crudus</em> in the Caribbean. <em>H. crudus</em> feeds abundantly on coconut leaves and population densities of <em>H. crudus</em> were observed to be up to 40 times higher in areas affected by palm lethal yellowing type syndromes than unaffected ones (Howard, 1980; Howard <em>et al</em>., 1983). Since then, evidence for <em>H. crudus</em> carrying phytoplasmas has been demonstrated (Dzido <em>et al</em>., 2020; Mou <em>et al</em>., 2022a; EPPO, 2024). The geographical distribution of <em>H</em><em>. crudus</em> in the Americas more or less coincides with the known distribution of the phytoplasmas (Howard, 1983). In areas other than the Americas, no vector has been positively identified. Other species of insects have been suggested as potential vectors (Brown <em>et al</em>., 2006; Bila <em>et al</em>., 2017; Fernández-Barrera <em>et al.,</em> 2022; Kwadjo <em>et al</em>., 2018; Lu <em>et al</em>., 2016; Mpunami <em>et al</em>., 2000; Ramos Hernández <em>et al.</em>, 2020).</p><p style="text-align: justify;">Purcell (1985) noted that <em>H. crudus</em> is a very inefficient vector of lethal yellowing, but can be so abundant that a very low transmission rate is sufficient to spread the disease.</p><p style="text-align: justify;">Oropeza <em>et al</em>. (2017) demonstrated the transmission of lethal yellowing phytoplasmas from coconut embryos to plantlets. However the relevance of seed transmission for disease spread is not clear.</p></div>
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-minus-circle" aria-hidden="true"></i> DETECTION AND IDENTIFICATION<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1" style="display: block;">
<h4><strong>Symptoms</strong></h4><h4 style="text-align: justify;"><strong>Morphology</strong></h4><p style="text-align: justify;">Typical phytoplasma particles were found in sieve tubes of infected plants. They were ovoid, elongated and filamentous in shape and were bounded by a triple-layered structure comprising two electron-dense layers with a transparent layer between them (Plavsic-Banjac <em>et al.,</em> 1972).</p><h4 style="text-align: justify;"><strong>Detection and inspection methods</strong></h4><p style="text-align: justify;">In general, an early symptom is the drying up of developing inflorescences, and the premature drop of most or all fruits or seeds within a few days. In coconut palms the spathes enclosing the flowers become discoloured and the tips blacken. The youngest leaves next to the buds show water-soaked streaks which spread until there is a terminal rot of the growing point. Following the first symptoms there is progressive leaf discoloration, beginning with the older leaves and spreading rapidly to the younger ones. The foliage turns light-yellow and eventually orange-yellow. This symptom coincides with the death of root tips. Death occurs in <em>C. nucifera</em> about 4 months after the initial symptoms appear.</p><p style="text-align: justify;">Several molecular methods exist to detect lethal yellowing type syndrome associated phytoplasmas both in plants and in vectors (Bahder <em>et al.</em>, 2018; Bahder <em>et al</em>., 2019; Christensen <em>et al</em>., 2004; Córdova <em>et al.</em>, 2014; Hodgetts <em>et al</em>., 2009; Lu <em>et al.</em>, 2016; Pilet, 2021). LAMP detection systems have also been developed to rapidly detect the phytoplasmas in the field (Dickinson, 2015; Tropicsafe, 2021a; Yu <em>et al</em>., 2023).</p><p style="text-align: justify;">Sampling is most reliable when done in immature leaves taken from around the apical meristem, or inflorescences, which are rich in phloem (Harrison <em>et al.</em>, 1999). However, once plants are symptomatic, PCR testing of the phloem from the palm trunk (by drilling a hole 10–15 cm into the trunk and collecting sawdust) is a non-destructive method of successful phytoplasma detection (Harrison <em>et al.</em>, 2002). The drill should be sterilised (e.g. in 70% ethanol) prior to each sample collection.</p></div>
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> PATHWAYS FOR MOVEMENT<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1">
<p style="text-align: justify;">Natural spread results from the movement of the known vector <em>H. crudus</em>, as well as of yet unknown vectors. Infected plants for planting, including ornamental species, can carry the pathogen in international trade. They can also carry infectious vectors, such as adults of <em>Haplaxius crudus</em> that are known to remain within palm foliage when plants are uprooted and transported to new localities. Ogle and Harries (2005) considered that the most likely means of transmission of the disease between Caribbean islands has been by the unintentional introduction of infected vectors on pasture grasses or animal fodder. Nymphs of <em>H. crudus</em> could also be moved in soil accompanying plants for planting or turf, but nymphs are not infected by the phytoplasmas.</p></div>
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> PEST SIGNIFICANCE<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1">
<h4><strong>Economic impact</strong></h4><p style="text-align: justify;">Palm lethal yellowing phytoplasmas are a serious economic threat for coconut, causing their premature decline and death. Experience in Jamaica suggests that almost total destruction of a population of susceptible palms can occur: by 1979, an estimated 4 million coconut palms had been killed by palm lethal yellowing disease on the island. In Florida, out of an estimated 1-1.5 million <em>C. nucifera</em> on the mainland, 300 000 had died by 1983. In the case of Florida, not a coconut-producing area, the socio-economic loss has been as a result of a destruction of an important and valuable feature of the amenity vegetation. The disease is still causing a large impact nowadays in the Caribbean and Central America (Gurr <em>et al</em>., 2016; Mou <em>et al.</em>, 2022b), as well as in Africa (Bertaccini <em>et al</em>., 2023; Tropicsafe 2021b; Mpunami <em>et al</em>., 1999) and Papua New Guinea (Miyazaki <em>et al.</em>, 2018).</p><h4 style="text-align: justify;"><strong>Control</strong></h4><p style="text-align: justify;">Management of the disease relies on several steps: early identification of infected plants (based on symptoms and ideally confirmed by testing), removal of infected plants, control of the vectors, planting of healthy plants, weed control.</p><p style="text-align: justify;">Research is being conducted to identify resistant genotypes (Gurr <em>et al</em>., 2016; Tropicsafe, 2021c).</p><p style="text-align: justify;">Control of the vector <em>H. crudus</em> with insecticides by spraying or trunk injection is used in some crops but has not been economically successful in coconut plantations (Gurr <em>et al</em>., 2016). Management of the grass populations in coconut plantations has also been suggested as a means of control, since some grass species are hosts for the nymphs of <em>H. crudus</em> (although they are relatively poor hosts) (Howard, 1990).</p><p style="text-align: justify;">Treatment of infected plants by injection into the trunk of antibiotics (e.g. oxytetracycline) had been shown to suppress the symptoms of lethal yellowing (McCoy, 1975) but was not considered sustainable as it should be applied at 4-monthly intervals and the use of antibiotics in agriculture is banned in many countries worldwide.</p><h4 style="text-align: justify;"><strong>Phytosanitary risk</strong></h4><p style="text-align: justify;">In the EPPO region, the most important economic crop at risk are date palms (<em>Phoenix dactylifera</em>) that are widely cultivated in Algeria, Morocco and Tunisia for fruit production. In addition, many palm species are used as ornamental trees in the entire EPPO region (indoor and outdoor). Movement of ornamental palms from infested areas could be just as hazardous to date-growing countries as the movement of date palms themselves, because of the possibility of spread from ornamental to date palms by existing auchenorrhynchan insects that may become local vectors.</p><p style="text-align: justify;">The disease introduction into the EPPO region could also impact ornamental palms of considerable amenity value grown outside in the southern EPPO region and as well as valuable collections of palms under glass in botanic gardens, etc. In addition, palms are of great significance in nearly all modern indoor landscape plantings and their value to the horticultural trade is considerable in particular for mature plants.</p></div>
<!-- -->
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> PHYTOSANITARY MEASURES<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1">
<p style="text-align: justify;">In order to prevent the introduction of the disease to the EPPO region, the importation of host plants for planting, originating from areas where the disease occurs should be prohibited. Alternatively, measures such as pest-free areas may be appropriate, as is the case for similar pests, as well as pest-free production sites or pest-free places of production with conditions also ensuring the absence of vectors. Healthy planting material of host plants can be produced in the framework of a certification scheme. Post-entry quarantine (including testing) could be used in special cases.</p></div>
<!-- -->
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> REFERENCES<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1">
<p>Arocha-Rosete Y, Diallo HA, Konan-Konan JL, Kouame AEP, Seka K, Kra KD, Toualy MN, Kwadjo KE, Daramcoum WAMP, Beugré NI, Ouattara BWM, Kouadjo Zaka CG, Allou K, Fursy-Rodelec ND, Doudjo-Ouattara ON, Yankey N, Dery S, Maharaj A, Saleh M, Summerbell R, Contaldo N, Paltrinieri S, Bertaccini A &amp; Scott J (2016) Detection and identification of the coconut lethal yellowing phytoplasma in weeds growing in coconut farms in Cote d’Ivoire. <em>Canadian Journal of Botany</em> <strong>38</strong>, 164–173</p><p>Bahder BW, Soto N, Komondy L, Mou DF, Humphries AR &amp; Helmick EE (2019) Detection and quantification of the 16SrIV-D phytoplasma in leaf tissue of common ornamental palm species in Florida using qPCR and dPCR. <em>Plant Disease</em> <strong>103</strong>(8), 1918-1922.</p><p>Bahder BW, Helmick EE, Mou DF, Harrison NA &amp; Davis R (2018) Digital PCR technology for detection of palm-infecting phytoplasmas belonging to group 16SrIV that occur in Florida. <em>Plant Disease</em> <strong>102</strong>(5), 1008-1014.</p><p>Beakbane AB, Slater CHW, &amp; Posnette AF (1972) Mycoplasmas in the phloem of coconut, <em>Cocos nucifera</em> L. with lethal yellowing disease. <em>Journal of Horticultural Science&nbsp;</em><strong>47</strong>, 256.</p><p>Bertaccini A, Arocha-Rosete Y, Contaldo N, Duduk B, Fiore N, Montano HG, Kube M, Kuo CH, Martini M, Oshima K &amp; Quaglino F (2022) Revision of the ‘<em>Candidatus</em> Phytoplasma’ species description guidelines. <em>International Journal of Systematic and Evolutionary Microbiology&nbsp;</em><strong>72</strong>(4), 005353.</p><p>Bertaccini A, Contaldo N, Feduzi G, Andeme AM, Yankey EN &amp; Rovesti L (2023) Molecular identification of ‘<em>Candidatus</em> Phytoplasma palmicola’ associated with coconut lethal yellowing in Equatorial Guinea. <em>Annals of Applied Biology</em> <strong>183</strong>(3), 262-270. <a href="https://doi.org/10.1111/aab.12854" rel="noopener noreferrer" target="_blank">https://doi.org/10.1111/aab.12854</a></p><p>Bila J, Mondjana A, Samils B, Högberg N, Wilson MR &amp; Santos L (2017) First report of ‘<em>Candidatus</em> Phytoplasma palmicola’ detection in the planthopper <em>Diostrombus mkurangai</em> in Mozambique. <em>Bulletin of Insectology</em> <strong>70</strong>(1), 45-48.</p><p>Brown SE, Been BO &amp; McLaughlin WA (2006) Detection and variability of the lethal yellowing group (16Sr IV) phytoplasmas in the <em>Cedusa</em> sp. (Hemiptera: Auchenorrhyncha: Derbidae) in Jamaica. <em>Annals of Applied Biology</em> <strong>149</strong>(1), 53-62.</p><p>Christensen NM, Nicolaisen M, Hansen M &amp; Schulz A (2004) Distribution of phytoplasmas in infected plants as revealed by real-time PCR and bioimaging. <em>Molecular Plant-Microbe Interactions</em> <strong>17</strong>(11), 1175-1184.</p><p>Córdova I, Oropeza C, Puch-Hau C, Harrison N, Collí-Rodríguez A, Narvaez M, Nic-Matos G, Reyes C &amp; Sáenz L (2014) A real-time PCR assay for detection of coconut lethal yellowing phytoplasmas of group 16SrIV subgroups A, D and E found in the Americas. <em>Journal of Plant Pathology</em> <strong>96</strong>(2), 343-352.</p><p>Dzido JL, Sánchez R, Dollet M, Julia JF, Narvaez M, Fabre S &amp; Oropeza C (2020) <em>Haplaxius crudus</em> (Hemiptera: Cixiidae) transmits the lethal yellowing phytoplasmas, 16SrIV, to <em>Pritchardia pacifica</em> Seem. &amp; H. Wendl (Arecaceae) in Yucatan, Mexico. <em>Neotropical Entomology</em> <strong>49</strong>, 795-805.</p><p>Dickinson M (2015) Loop-mediated isothermal amplification (LAMP) for detection of phytoplasmas in the field. In: Lacomme C (ed.). Plant Pathology: Techniques and Protocols, Vol. 1302, Humana Press, New York, NY, USA. pp. 99–112.</p><p>EFSA (2017) Scientific Opinion on pest categorisation of Palm lethal yellowing phytoplasmas. <em>EFSA Journal&nbsp;</em><strong>15</strong>(10), 5028, 27 pp. <a href="https://doi.org/10.2903/j.efsa.2017.5028" rel="noopener noreferrer" target="_blank">https://doi.org/10.2903/j.efsa.2017.5028</a></p><p>EPPO (2024) <em>Haplaxius crudus</em>. EPPO datasheets on pests recommended for regulation. <a href="https://gd.eppo.int">https://gd.eppo.int</a> (accessed 2024-02-06).</p><p>Fernández-Barrera M, Córdova-Lara I, Chan-Rodríguez JL, Castillo-Vera A, Blanco-Rodríguez E, Nic-Matos G, Oropeza-Salín C &amp; Sáenz-Carbonell L (2022) Detection of 16SrIV-A phytoplasma DNA in <em>Colpoptera</em> sp. (Hemiptera: Nogodinidae) insects in Yucatan Peninsula, Mexico. <em>Brazilian Journal of Biology&nbsp;</em><strong>84</strong>, Article e257470. <a href="https://doi.org/10.1590/1519-6984.25" rel="noopener noreferrer" target="_blank">https://doi.org/10.1590/1519-6984.25</a></p><p>Gurr GM, Johnson AC, Ash GJ, Wilson BA, Ero MM, Pilotti CA, Dewhurst CF &amp; You MS (2016) Coconut lethal yellowing diseases: a phytoplasma threat to palms of global economic and social significance. <em>Frontiers in plant science</em> <strong>7</strong>, 1521. <a href="https://doi.org/10.3389/fpls.2016.01521" rel="noopener noreferrer" target="_blank">https://doi.org/10.3389/fpls.2016.01521</a></p><p>Harrison NA, Womack M &amp; Carpio ML (2002) Detection and characterization of a lethal yellowing (16SrIV) group phytoplasma in Canary Island date palms affected by lethal decline in Texas. <em>Plant Disease&nbsp;</em><strong>86</strong>(6), 676-681.</p><p>Hodgetts J, Boonham N, Mumford R &amp; Dickinson M (2009) Panel of 23S rRNA gene-based real-time PCR assays for improved universal and group-specific detection of phytoplasmas. <em>Applied and Environmental Microbiology</em> <strong>75</strong>(9), 2945-2950.</p><p>Howard FW (1980) Population densities of <em>Myndus crudus</em> Van Duzee (Homoptera: Cixiidae) in relation to coconut lethal yellowing distribution in Florida. <em>Principes&nbsp;</em><strong>24</strong>, 174-178.</p><p>Howard FW, Norris R &amp; Thomas D (1983) Evidence of transmission of palm lethal yellowing agent by a planthopper, <em>Myndus crudus</em> (Homoptera, Cixiidae). <em>Tropical Agriculture</em> <strong>60</strong>, 168–171.</p><p>Howard FW (1983) World distribution and possible geographical origin of palm lethal yellowing disease and its vectors. <em>FAO Plant Protection Bulletin</em> <strong>31</strong>, 101-113.</p><p>Howard FW (1990) Evaluation of grasses for cultural control of <em>Myndus crudus,</em> a vector of lethal yellowing of palms. <em>Entomologia Experimentalis et Applicata</em> <strong>56</strong>, 131-137.</p><p>Howard FW, Thomas DL, Donselman HM &amp; Collins ME (1979) Susceptibilities of palm species to mycoplasma-like organism-associated diseases in Florida. <em>FAO Plant Protection Bulletin</em> <strong>27</strong>, 109-117.</p><p>Jones LM, Pease B, Perkiins SL, Constable FE, Kinoti WM, Warmington D, Allgood B, Powell S, Taylor P, Pearce C &amp; Davis RI (2021) ‘<em>Candidatus</em> Phytoplasma dypsidis’, a novel taxon associated with a lethal wilt disease of palms in Australia. <em>International Journal of Systematic and Evolutionary Microbiology</em> <strong>71</strong>(5). <a href="https://doi.org/10.1099/ijsem.0.004818" rel="noopener noreferrer" target="_blank">https://doi.org/10.1099/ijsem.0.004818</a></p><p>Lu H, Wilson BA, Ash GJ, Woruba SB, Fletcher MJ, You M, Yang G &amp; Gurr GM (2016) Determining putative vectors of the Bogia Coconut Syndrome phytoplasma using loop-mediated isothermal amplification of single-insect feeding media. <em>Scientific Reports&nbsp;</em><strong>6</strong>(1), 35801.</p><p>McCoy RE (1975) Effect of oxytetracycline dose and stage of development on remission of lethal yellowing in coconut palm. <em>Plant Disease Reporter</em> <strong>59</strong>, 717-720.</p><p>Miyazaki A, Shigaki T, Koinuma H, Iwabuchi N, Rauka GB, Kembu A, Saul J, Watanabe K, Nijo T, Maejima K &amp; Yamaji Y (2018) ‘<em>Candidatus&nbsp;</em>Phytoplasma noviguineense’, a novel taxon associated with Bogia coconut syndrome and banana wilt disease on the island of New Guinea. <em>International Journal of Systematic and Evolutionary Microbiology</em> <strong>68</strong>(1), 170-175.</p><p>Mou DF, Di Lella B, Halbert SE, Bextine B, Helmick EE &amp; Bahder BW (2022a) Acquisition and transmission of the lethal bronzing phytoplasma by <em>Haplaxius crudus</em> using infected palm spear leaves and artificial feeding media<em>. Phytopathology</em> <strong>112</strong>(10), 2052-2061.</p><p>Mou DF, Helmick EE &amp; Bahder BW (2022b) Multilocus sequence analysis reveals new hosts of palm lethal decline phytoplasmas in Florida, USA. <em>Plant Health Progress</em> <strong>23</strong>(4), 399-408.</p><p>Mpunami AA, Tymon A &amp; Dickinson MJ (1999) Genetic diversity in the coconut lethal yellowing disease phytoplasmas of East Africa. <em>Plant Pathology</em> <strong>48</strong>(1), 109-114.</p><p>Mpunami A, Tymon A, Jones P &amp; Dickinson MJ (2000) Identification of potential vectors of the coconut lethal disease phytoplasma. <em>Plant Pathology</em> <strong>49</strong>(3), 355-361.</p><p>Ogle L &amp; Harries H (2005) Introducing the vector: how coconut lethal yellowing disease may have reached the Caribbean. <em>Ethnobotany Research and Applications</em> <strong>3</strong>, 139–142</p><p>Oropeza C, Cordova I, Puch‐Hau C, Castillo R, Chan JL &amp; Sáenz L (2017) Detection of lethal yellowing phytoplasma in coconut plantlets obtained through in vitro germination of zygotic embryos from the seeds of infected palms. <em>Annals of Applied Biology</em> <strong>171</strong>(1), 28-36.</p><p>Palma-Cancino PJ, Ramos-Hernández E &amp; Ortiz-García CF (2023) History of lethal yellowing with emphasis on the susceptibility of royal palms (<em>Roystonea</em> spp.). <em>Agro Productividad.</em> <a href="https://doi.org/10.32854/agrop.v16i4.2488" rel="noopener noreferrer" target="_blank">https://doi.org/10.32854/agrop.v16i4.2488</a></p><p>Pilet F (2021) Diagnostic for Phytoplasmae associated with coconut lethal yellowing type syndromes from Africa, and elsewhere. Cirad Webinar (2021-01-21-21).</p><p>Plavsic-Banjac B, Hunt P &amp; Maramorosch K (1972) Mycoplasma-like bodies associated with lethal yellowing disease of coconut palms. <em>Phytopathology&nbsp;</em><strong>62</strong>, 298-299.</p><p>Purcell AH (1985) The ecology of plant diseases spread by leafhoppers and planthoppers. In: <em>The leafhoppers and planthoppers</em> (Ed. by Nault, L.R.; Rodriguez, J.G.), pp. 351-380. Wiley, New York, USA.</p><p>Ramos Hernández E, Lesher Gordillo MJ, Ortiz García CF, Oropeza Salín C, Sánchez Soto S, Magaña Alejandro MÁ &amp; Narváez MD (2020) Cixiidae and Derbidae (Hemiptera) putative vectors of phytoplasma of group 16 SrIV. <em>Revista Colombiana De Entomología</em> <strong>46</strong>(2), e7065. <a href="https://doi.org/10.25100/socolen.v46i2.7065" rel="noopener noreferrer" target="_blank">https://doi.org/10.25100/socolen.v46i2.7065</a></p><p>Soto N, Helmick EE, Harrison NA &amp; Bahder BW (2021) Genetic variability of palm lethal decline phytoplasmas in the Caribbean basin and Florida, USA, based on a multilocus analysis. <em>Phytopathology </em><strong>111</strong>(12), 2203-2212. <a href="https://doi.org/10.1094/PHYTO-04-21-0130-R" rel="noopener noreferrer" target="_blank">https://doi.org/10.1094/PHYTO-04-21-0130-R</a></p><p>Tropicsafe (2021a) Rapid infield detection of coconut phytoplasma in Africa. Final Handbook with innovation Factsheets. <a href="http://www.tropicsafe.eu/wp-content/uploads/2022/02/RAPID-IN-FIELD-DETECTION-OF-COCO-PHYTO-AF.pdf" rel="noopener noreferrer" target="_blank">http://www.tropicsafe.eu/wp-content/uploads/2022/02/RAPID-IN-FIELD-DETECTION-OF-COCO-PHYTO-AF.pdf</a></p><p>Tropicsafe (2021b) Coconut sector: market analysis and socioeconomic aspects -A general framework of the coconut sector in Jamaica, Ghana, and Mexico. Final Handbook with innovation Factsheets. <a href="http://www.tropicsafe.eu/wp-content/uploads/2022/02/COCONUT-SECTOR-MARKET.pdf" rel="noopener noreferrer" target="_blank">http://www.tropicsafe.eu/wp-content/uploads/2022/02/COCONUT-SECTOR-MARKET.pdf</a></p><p>Tropicsafe (2021c) Coconut lethal yellowing resistance screening and disease management - The use of lethal yellowing resistant germplasm. Final Handbook with innovation Factsheets. <a href="https://www.tropicsafe.eu/wp-content/uploads/2022/02/COCONUT-LETHAL-YELLOWING-RESISTANCE.pdf" rel="noopener noreferrer" target="_blank">https://www.tropicsafe.eu/wp-content/uploads/2022/02/COCONUT-LETHAL-YELLOWING-RESISTANCE.pdf</a></p><p>Wei W, Davis RE, Lee IM &amp; Zhao Y (2007) Computer-simulated RFLP analysis of 16S rRNA genes: identification of ten new phytoplasma groups. <em>International Journal of Systematic and Evolutionary Microbiology&nbsp;</em><strong>57</strong>(8), 1855-1867. <a href="https://doi.org/10.1099/ijs.0.65000-0" rel="noopener noreferrer" target="_blank">https://doi.org/10.1099/ijs.0.65000-0</a></p><p>Yu SS, Pan YW, Zhu H &amp; Song WW (2023) Universal, rapid, and visual detection methods for phytoplasmas associated with coconut lethal yellowing diseases targeting 16S rRNA gene sequences. <em>Plant Disease</em> <strong>107</strong>(2), 276-280.</p></div>
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> ACKNOWLEDGEMENTS<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1">
<p>This datasheet was extensively revised in 2024 by the EPPO Secretariat with the help of Fabian Pilet (CIRAD). His valuable contribution is gratefully acknowledged.</p></div>
<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> How to cite this datasheet?</h3></a>
<div class="para1">EPPO (2024) <i>Palm lethal yellowing type syndromes</i>. EPPO datasheets on pests recommended for regulation. <a href="https://gd.eppo.int" target="_blank">https://gd.eppo.int</a> (accessed 2024-11-25)
</div>

<!-- -->
<a class="btntitle" href="#"><h3><i class="fa fa-plus-circle" aria-hidden="true"></i> Datasheet history<small class="pull-right" title="Last modification"><i class="fa fa-clock-o" aria-hidden="true"></i> 2024-06-18</small></h3></a>
<div class="para1">
<p style="text-align: justify;">This datasheet was first published in the EPPO Bulletin in 1986 and revised in the two editions of 'Quarantine Pests for Europe' in 1992 and 1997. It is now maintained in an electronic format in the EPPO Global Database. The sections on 'Identity', ‘Hosts’, and 'Geographical distribution' are automatically updated from the database. For other sections, the date of last revision is indicated on the right.</p><p>CABI/EPPO (1992/1997) Quarantine Pests for Europe (1st and 2nd edition). CABI, Wallingford (GB).</p><p>EPPO (1986) Data sheets on quarantine organisms No. 159,<em>&nbsp;Palm lethal yellowing mycoplasm. EPPO Bulletin</em> <strong>16</strong>, 61-66.&nbsp;<a href="https://onlinelibrary.wiley.com/doi/epdf/10.1111/j.1365-2338.1986.tb01138.x" rel="noopener noreferrer" target="_blank">https://onlinelibrary.wiley.com/doi/epdf/10.1111/j.1365-2338.1986.tb01138.x</a></p></div>
<!-- -->
    <img src="/media/img/eu_funding_250.png">

"""
ignore_list = ['references', 'acknowledgements', 'how to cite this datasheet?', 'datasheet history']
from tqdm import tqdm
def get_datasheet(url):
    import requests
    from bs4 import BeautifulSoup
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    datasheet = soup.find('div', class_='datasheet')
    if datasheet is None:
        return None
    datasheet_dict = {}
    for btn in datasheet.find_all('h3'):
        # remove 'pull-right' class from h3 children
        for child in btn.find_all('small'):
            child.decompose()        
        key = btn.text.lower()
        if key in ignore_list:
            continue
        datasheet_dict[key] = {}
        if btn.find_next('div', class_='para0'):
            datasheet_dict[key]['description'] = btn.find_next('div', class_='para0').text.replace('\n', ' ')
        elif btn.find_next('div', class_='para1'):
            if btn.find_next('div', class_='para1').find_next().name == 'p':
                datasheet_dict[key]['description'] = btn.find_next('div', class_='para1').text.replace('\n', ' ')
            elif btn.find_next('div', class_='para1').find_next().name == 'h4':
                for h4 in btn.find_next('div', class_='para1').find_all('h4'):
                    key1 = h4.text.lower()
                    datasheet_dict[key][key1] = h4.find_next('p').text.replace('\n', ' ')      
        
    return datasheet_dict

    
```

```python
# new dict copy with keys without \u00a0
import unicodedata
def remove_non_breaking_space(cist):
    new_cist = {}
    for key, value in cist.items():
        if isinstance(value, dict):
            new_cist[unicodedata.normalize('NFKD', key).encode('ascii', 'ignore').decode('ascii')] = remove_non_breaking_space(value)
        elif isinstance(value, str):
            new_cist[unicodedata.normalize('NFKD', key).encode('ascii', 'ignore').decode('ascii')] = unicodedata.normalize('NFKD', value).encode('ascii', 'ignore').decode('ascii').replace('\"', "'").replace("\r",' ')
    return new_cist

datasheet_dict = remove_non_breaking_space(datasheet_dict)
```
